```python
import pandas as pd

df = pd.read_csv('../data/01_input_history.csv')

print(df.head())

prodotti = df['Product'].unique()
print(len(prodotti))

paesi = df['Country'].unique()
print(len(paesi))
```

```python
from matplotlib import pyplot as plt
import random

def random_color():
    return (random.uniform(0, 1), random.uniform(0, 1), random.uniform(0, 1))

def country_plot(df, country):
    products = df["Product"].unique()
    months = df["Month"].unique()
    for prod in products:
        plt.plot(months, df.loc[((df["Country"] == country) & (df["Product"] == prod)), ["Quantity"]], c = random_color())
    plt.title(f"{country} sales over time")
    plt.show()

def year_plot(df, country, product):
    years = [i for i in range (2004, 2025)]
    for y in years:
        months = df.loc[((df["Country"] == country) & (df["Product"] == product) & (df["Month"].apply(lambda x : x[3:]) == str(y))), ["Month"]]
        quantities = df.loc[((df["Country"] == country) & (df["Product"] == product) & (df["Month"].apply(lambda x : x[3:]) == str(y))), ["Quantity"]]
        plt.plot(months["Month"].apply(lambda x : x[:3]), quantities["Quantity"].apply(lambda x : int(x)), c = random_color(), label=f"{y}")
    plt.legend()
    plt.show()
```

```python
country = "Australia"
product = "GentleGlow Baby Face Wash"
year_plot(df, country, product)
```
